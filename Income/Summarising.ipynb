```python
employee = pd.read_excel('testData.xlsx', sheet_name= 'employee')
workday = pd.read_excel('testData.xlsx', sheet_name= 'workday')

print(workday.shape)
print(employee.shape)
```

```python
data1 = workday.join(employee, lsuffix="_l", rsuffix="_r", on='employee_id' , how = "left")

data1.shape
```

```python
# data1.info()

data1 = data1.drop(['employee_id_r'], axis=1)

data1 = data1.rename(columns = {'employee_id_l':'employee_id'})

data1.info()

# data2 = workday.merge(employee, how='left', on='employee_id' )
# data2.shape
```

```python
# Obtener las estadística para cada variable

data1.describe()
```

```python
# Obtener, por empleado, las ausencias, ordenar de mayor a menor por las ausencias

ausencias = data1.groupby(['employee_id']).absent.sum().reset_index().sort_values(by=['absent'], ascending=True)
ausencias

```

```python
# Obtener, por mes, las ausencias, ordenar de mayor a menor por las ausencias

mes = data1.groupby(['month']).absent.sum().reset_index().sort_values(by=['absent'], ascending=True)
mes
```

```python
# Obtener, por empleado, las faltas cometidas

faltas = data1.groupby(['employee_id']).disciplinary_faults.sum().reset_index().sort_values(by=['disciplinary_faults'], ascending=False)
faltas
```

```python
# Mostrar por empleado, los datos promedios de transporte 

Trasnporte = data1.groupby(['employee_id']).transport_expense.sum().reset_index().sort_values(by=['transport_expense'], ascending=False)
Trasnporte
```

```python
# Agrupar por Género e hijos, las suma de las ausencias

gen_hijos_ausen = data1.groupby(['gender','sons']).agg({'absent':'sum',
                                                             'disciplinary_faults':'sum'
                                                                }).reset_index()

gen_hijos_ausen
```

```python
# Agrupar transportation_method y genero , los gasstos mínmos, máximos y promedios de transport_expense, para edades superiores a los 35 años

trasmet_genero_edad35 = data1[data1["age"]>=36].groupby(['transportation_method','gender']).agg({'transport_expense':'min'
                                                                                               }).reset_index().sort_values(by=['transport_expense'], ascending=False)
trasmet_genero_edad35
```

```python
# Agrupar transportation_method y genero , los gasstos mínmos, máximos y promedios de transport_expense, para edades superiores a los 35 años y con más de 1 en educacion

trasmet_genero_edad35_edumas1 = data1[(data1.age >= 35) & (data1.education > 1)].groupby(['transportation_method','gender']).agg({'transport_expense':'min'
                                                                                               }).reset_index().sort_values(by=['transport_expense'], ascending=False)
trasmet_genero_edad35_edumas1
```
