```python
#####
# 1 #  
##################################################################################################
##################################################################################################
##################################################################################################
#                              Importación de los ingresos                                       #
##################################################################################################
##################################################################################################
##################################################################################################

#############################
#   Ingresos  2007 - 2020   #
#############################

ingresos_2007_2021 = pd.read_excel('SIGAF_2007_2021.xlsx') 
# print(ingresos_2007_2021.shape)
# print(ingresos_2007_2021.columns)


###############################
#   Ingresos  SIGAF  actual   #
###############################

# Este si se debe llamar siempre "SIGAF_Ingresos_actual."

SIGAF_Ingresos_actual = pd.read_excel('SIGAF_Ingresos_actual.xlsx') 

SIGAF_Ingresos_actual['PosPre'] = SIGAF_Ingresos_actual['PosPre'].replace(['I'],'I0000000000000')
# print(SIGAF_Ingresos_actual.shape)
# print(SIGAF_Ingresos_actual.columns)

###############################################
#   Presupuesto inicial   #
###############################################

PPI = pd.read_excel('PPI.xlsx', converters={
                                            'Año':str,
                                            'Clase':str,
                                            'Subclase':str,
                                            'Grupo':str,
                                            'Subgrupo':str,
                                            'Partida':str,
                                            'Subpartida':str,
                                            'Renglón':str,
                                            'Subrenglón':str,
                                            'Fuente finaciamiento':str,
                                            'Descripción':str,
                                            'Presupuesto inicial': float
                                            }
                    ) 
# print(PPI.shape)
# print(PPI.columns)
```

```python
PPI['Pospreanual'] = PPI[['Año','Clase','Subclase','Grupo','Subgrupo','Partida','Subpartida','Renglón','Subrenglón','Fuente finaciamiento']].agg(''.join, axis=1)
PPI_a = PPI.loc [:, ['Pospreanual','Presupuesto inicial']]
PPI_a
```

```python
##################################################################################################
##################################################################################################
##################################################################################################
#                                  Ingresos  Anuales                                             #
##################################################################################################
##################################################################################################
##################################################################################################

#############################
#   Ingresos  20167-2020     #
#############################

ingresos_2007_2021_a = ingresos_2007_2021 
# ingresos_2007_2021_a.info()

# Año = today.year

SIGAF_Ingresos_actual_a = SIGAF_Ingresos_actual
SIGAF_Ingresos_actual_a['Año'] = Año



# SIGAF_Ingresos_actual_a.info()

# ingresos_2007_2021_a.columns = SIGAF_Ingresos_actual_a.columns


# Ingresos_a = pd.merge(ingresos_2007_2021_a, SIGAF_Ingresos_actual_a, on='name')

Ingresos_a = pd.concat([ingresos_2007_2021_a, SIGAF_Ingresos_actual_a], axis=0, ignore_index=True )

convert_dict_2 = {'Año': str
                }

Ingresos_a = Ingresos_a.astype(convert_dict_2)

print(Ingresos_a.shape)
Ingresos_a
# Ingresos_a.tail(15)
```

```python
# Ingresos_a.info()
```

```python
###########################################################
#       Transformación de datos                           #
###########################################################

Ingresos_a['Pospre']         = Ingresos_a['PosPre'].str.slice(1,14)
Ingresos_a['Pospreanual']    = Ingresos_a[['Año','Pospre']].agg(''.join, axis=1)
Ingresos_a = Ingresos_a.reindex(columns=['Año', 'Nivel', 'n_Nivel' ,'PosPre','Pospre','Pospreanual', 'cod_sector','Presupuesto actual', 'mes.cod','mes','cod_clase','clase','cod_subclase','subclase','cod_grupo','grupo','cod_subgrupo','subgrupo',
         'cod_partida','cod_subpartida','cod_renglon','cod_subrenglon', 'cod_fuentefinanciacion','cod_fuentefinanciacion_3','Descripción','Ingresos'])
# Ingresos_a.info()
Ingresos_a.shape

```

```python

Ingresos_a_merge = pd.merge(Ingresos_a, PPI_a, how='left', on='Pospreanual')
Ingresos_a_merge.shape
```
